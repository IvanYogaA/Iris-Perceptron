```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
NIM = "622021004"
file_name = f"Percobaan3_622021004.npz"
data = np.load(file_name)
```

```python
setosa_data = data['setosa']
versicolor_data = data['versicolor']
X_setosa = setosa_data[:, :2]
X_versicolor = versicolor_data[:, :2]
```

```python
target_setosa = np.zeros(len(X_setosa))
target_versicolor = np.ones(len(X_versicolor))
X_combined = np.vstack((X_setosa, X_versicolor))
targets = np.hstack((target_setosa, target_versicolor))
```

```python
parameter_vector = np.zeros(X_combined.shape[1])
threshold = 0
learning_step = 0.1
iteration_count = 20

def decision_function(value):
    return 1 if value >= 0 else 0
```

```python
for _ in range(iteration_count):
    for index in range(len(X_combined)):
        model_output = decision_function(np.dot(parameter_vector, X_combined[index]) + threshold)
        error_value = targets[index] - model_output
        parameter_vector += learning_step * error_value * X_combined[index]
        threshold += learning_step * error_value

print("Final Parameter:", parameter_vector)
print("Final Threshold:", threshold)
```

```python
plt.figure(figsize=(10, 6))
for i, (point, label) in enumerate(zip(X_combined, targets)):
    color = 'red' if label == 0 else 'blue'
    label_text = 'Setosa' if label == 0 and i == 0 else 'Versicolor' if label == 1 and i == len(X_setosa) else ""
    plt.scatter(point[0], point[1], color=color, label=label_text)
```

```python
boundary_x_vals = np.linspace(min(X_combined[:, 0]), max(X_combined[:, 0]), 100)
boundary_y_vals = -(parameter_vector[0] * boundary_x_vals + threshold) / parameter_vector[1]
plt.plot(boundary_x_vals, boundary_y_vals, color='black', label='Decision Boundary')

plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.title(f"Perceptron Model (NIM: {NIM})")
plt.legend()
plt.grid(True)
plt.show()
```
